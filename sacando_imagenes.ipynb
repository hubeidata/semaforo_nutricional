```python
import requests
import json
import os
import urllib.parse

# Configuración de la API
api_key = 'AIzaSyABKEYVWlT_gmaedRwXQsr4e8jhd3HT3XMT' # Reemplaza con tu API key
search_engine_id = 'f37049d0eaf194cb1A'  # Reemplaza con tu Search Engine ID
num_results = 150  # Número total de imágenes a descargar por plato

# Lista de platos
lista_platos = ["soltero de queso"]

# Función para descargar imágenes
def descargar_imagenes(plato, num_results):
    query = urllib.parse.quote(plato)
    path = f'/content/drive/My Drive/semaforo/Descargas/{plato}'
    if not os.path.exists(path):
        os.makedirs(path)

    results_per_page = 10
    for start_index in range(1, num_results + 1, results_per_page):
        url = f"https://www.googleapis.com/customsearch/v1?q={query}&cx={search_engine_id}&key={api_key}&searchType=image&num={results_per_page}&start={start_index}"
        response = requests.get(url)
        results = response.json()

        # Descargar las imágenes
        for i, item in enumerate(results.get('items', [])):
            try:
                img_url = item['link']
                img_data = requests.get(img_url).content
                with open(os.path.join(path, f'{plato}_{start_index + i}.jpg'), 'wb') as handler:
                    handler.write(img_data)
                print(f'{plato}: Imagen {start_index + i} descargada de {img_url}')
            except Exception as e:
                print(f'Error descargando {plato} imagen {start_index + i}: {e}')

# Iterar sobre la lista de platos y descargar imágenes
for plato in lista_platos:
    descargar_imagenes(plato, num_results)
```
