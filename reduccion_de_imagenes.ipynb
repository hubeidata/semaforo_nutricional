```python
import os
from PIL import Image

# Directorio de entrada y salida
directorio_entrada = "/content/drive/MyDrive/semaforo/Ajì_de_Gallina "
directorio_salida = "/content/drive/MyDrive/semaforo/reducidas"
# Crear el directorio de salida si no existe
if not os.path.exists(directorio_salida):
    os.makedirs(directorio_salida)

# Definir el área a recortar: (left, upper, right, lower)
area = (100, 100, 400, 400)

# Iterar sobre los archivos en el directorio de entrada
for nombre_archivo in os.listdir(directorio_entrada):
    if nombre_archivo.endswith((".jpg", ".jpeg", ".png", ".bmp", ".gif")):
        # Ruta completa del archivo de entrada
        ruta_archivo = os.path.join(directorio_entrada, nombre_archivo)

        # Abrir la imagen
        imagen = Image.open(ruta_archivo)

        # reducir la imagen
        imagen_recortada = imagen.crop(area)

        # Ruta completa del archivo de salida
        ruta_archivo_salida = os.path.join(directorio_salida, nombre_archivo)

        # Guardar la imagen recortada
        imagen_recortada.save(ruta_archivo_salida)

        print(f"Imagen recortada guardada en: {ruta_archivo_salida}")

print("Proceso de recorte completado.")
```

reducir imagenes


```python
import cv2
import os

# Definir rutas
ruta_entrada = "/content/drive/MyDrive/semaforo/Ajì_de_Gallina "
ruta_salida = "/content/drive/MyDrive/semaforo/reducidas"

# Recorrer la carpeta de entrada
for imagen in os.listdir(ruta_entrada):
    # Leer la imagen actual
    imagen_actual = cv2.imread(os.path.join(ruta_entrada, imagen))

    # Redimensionar la imagen
    imagen_redimensionada = cv2.resize(imagen_actual, (150, 150))

    # Guardar la imagen redimensionada
    cv2.imwrite(os.path.join(ruta_salida, imagen), imagen_redimensionada)
```

```python
import cv2
import os

# Definir la ruta de entrada
ruta_entrada = "/content/drive/MyDrive/semaforo/comidas_platos/validation"  # Reemplazar con la ruta real

# Definir la ruta de salida
ruta_salida = "/content/drive/MyDrive/semaforo/comidas2_platos2/validation" # Reemplazar con la ruta real

# Recorrer las carpetas en la ruta de entrada
for carpeta in os.listdir(ruta_entrada):
    ruta_carpeta_entrada = os.path.join(ruta_entrada, carpeta)
    ruta_carpeta_salida = os.path.join(ruta_salida, carpeta)

    if not os.path.exists(ruta_carpeta_salida):
        os.makedirs(ruta_carpeta_salida)  # Crear la carpeta de salida si no existe

    # Recorrer las imágenes en la carpeta actual
    for imagen in os.listdir(ruta_carpeta_entrada):
        ruta_imagen_entrada = os.path.join(ruta_carpeta_entrada, imagen)
        ruta_imagen_salida = os.path.join(ruta_carpeta_salida, imagen)

        # Leer la imagen
        imagen_original = cv2.imread(ruta_imagen_entrada)

        # Redimensionar la imagen a 200 x 200
        imagen_redimensionada = cv2.resize(imagen_original, (200,200))

        # Guardar la imagen redimensionada
        cv2.imwrite(ruta_imagen_salida, imagen_redimensionada)

        print(f"Imagen redimensionada: {ruta_imagen_salida}")
```
